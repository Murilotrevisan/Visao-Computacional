### 2) Mudanças na resolução de escala de cinza

1. Agora vamos reduzir a escala de níveis de cinza da imagem. Para isso vamos criar uma função de nome ```transformaImg```.  Já disponibilizamos o escopo da função e uma descrição das entradas e saídas. Comente o que cada linha da função está exercendo. 

2. Agora utilize a mesma e reduza a escala de níveis de cinza da imagem ```angiography.tif``` de 256 níveis para:

  - 64
  - 32
  - 16
  - 8
  - 4
  - 2

3. Mostrar todas as imagens (no mesmo tamanho) e comentar os resultados.

Obs: Comente os resultados!

````python
import numpy as np

# função definida que tem como entradas a imagem e o nivel de cinza desejado
def transformaImg(img, grayLevel): 
  '''
  Descrição: Transforma uma imagem para uma nova escala de cinza.

  Entrada:
    - img: Imagem de entrada (uint8).
    - grayLevel: Nova escala de níveis de cinza desejada (uint8).
  
  Saída:
    - res: Imagem com nova escala de cinza (uint8).
  '''
# converte a imagem de entrada em um float de 32 bits
  img_in = img.copy().astype('float32') 
# cria uma variavel que recebe o nivel de cinza da imagem e divide por 255 e arredonda o resultado
  res = np.round( img_in * (grayLevel - 1) / 255.); 
# converte novamente a imagem em um inteiro
  res = res.astype('uint8')
# retorna como saida da função a nova imagem com a resolução desejada
  return res

## -- TODO -- ##
# Leitura da imagem
angio = cv.imread("angiography.tif")
# impressão da imagem
plt.figure(figsize = (5,5))
plt.imshow(angio)

# mudança na escala de cinza para 64
angio64 = transformaImg(angio, 64)
# impressão da nova imagem
plt.figure(figsize = (5,5))
plt.imshow(angio64)

# mudança na escala de cinza para 32
angio32 = transformaImg(angio, 32)
# impressão da nova imagem
plt.figure(figsize = (5,5))
plt.imshow(angio32)

# mudança na escala de cinza para 16
angio16 = transformaImg(angio, 16)
# impressão da nova imagem
plt.figure(figsize = (5,5))
plt.imshow(angio16)

# mudança na escala de cinza para 8
angio8 = transformaImg(angio, 8)
# impressão da nova imagem
plt.figure(figsize = (5,5))
plt.imshow(angio8)

# mudança na escala de cinza para 4
angio4 = transformaImg(angio, 4)
# impressão da nova imagem
plt.figure(figsize = (5,5))
plt.imshow(angio4)

# mudança na escala de cinza para 2
angio2 = transformaImg(angio, 2)
# impressão da nova imagem
plt.figure(figsize = (5,5))
plt.imshow(angio2)
````

Obs: Comente os resultados!

````python
## -- TODO -- ##

regua = cv.imread("ruler.jpg")
plt.figure(figsize = (5,5))
plt.imshow(regua)
print(regua.shape)
# a imagem possui 770 pontos em y e a regua possui 8.5 cm
# portanto, podemos calcular qual o tamanho de um pixel em cm
# se 770 pixels equivalem a 8.5 cm, temos que um pixel tem em y a dimensão de:
[x,y,z] = regua.shape
print("a regua possui em y %d pixels" % y)
# efetuando a regra de tres
pix_measure = (85/770)
print("o tamanho em y de cada pixel é de %f mm" % pix_measure)
# dado o tamanho do pixel em mm
# podemos calcular quantos pixels há em uma polegada
DPI = (25.4/pix_measure)
# tendo o valor em DPI arredondamos para o menor inteiro
# e assim temos o DPI do scan
print("o scanner possui %d DPI " % np.floor(DPI))
````
